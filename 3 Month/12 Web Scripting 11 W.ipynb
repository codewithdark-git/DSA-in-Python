```python
final=pd.DataFrame()
for j in range(1,1001):
  webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page={}'.format(j)).text
  soup=BeautifulSoup(webpage,'lxml')
  company=soup.find_all('div',class_='company-content-wrapper')
  name=[]
  rating=[]
  reviews=[]
  ctype=[]
  hq=[]
  how_old=[]
  no_of_employee=[]

  for i in company:

    try:
       name.append(i.find('h2').text.strip())
    except:
       name.append(np.nan)

    try:
       rating.append(i.find('p',class_='rating').text.strip())
    except:
       rating.append(np.nan)
   
    try:

      reviews.append(i.find('a' , class_='review-count').text.strip())
    except:
      reviews.append(np.nan)

    try:

      ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
    except:
      ctype.append(np.nan)
    try:

      hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
    except:
      hq.append(np.nan)
    
    try:

      how_old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    except:
      how_old.append(np.nan)
    try:
      no_of_employee.append(i.find_all('p',class_='infoEntity')[3].text.strip())
    except:
      no_of_employee.append(np.nan)
    

  df=pd.DataFrame({'name':name,
    'rating':rating,
    'reviews':reviews,
    'company_type':ctype,
    'Head_Quarters':hq,
    'Company_Age':how_old,
    'No_of_Employee':no_of_employee,
    })
  
  final=final.append(df,ignore_index=True)



```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import random

# Function to extract job data from a single page
def extract_jobs_from_page(soup):
    jobs = []
    
    # Find all job containers
    for job_card in soup.find_all('div', class_='job_seen_beacon'):
        job_title = job_card.find('h2', class_='jobTitle').text.strip()
        company = job_card.find('span', class_='companyName')
        company = company.text.strip() if company else 'N/A'
        location = job_card.find('div', class_='companyLocation')
        location = location.text.strip() if location else 'N/A'
        salary = job_card.find('div', class_='salary-snippet')
        salary = salary.text.strip() if salary else 'N/A'
        job_description = job_card.find('div', class_='job-snippet')
        job_description = job_description.text.strip() if job_description else 'N/A'
        
        jobs.append({
            'Job Title': job_title,
            'Company': company,
            'Location': location,
            'Salary': salary,
            'Job Description': job_description
        })
    return jobs

# Main function to fetch multiple pages
def fetch_job_listings(base_url, pages=1):
    all_jobs = []
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.83 Safari/537.36'
    }
    
    for page in range(0, pages):
        url = f"{base_url}&start={page*10}"
        session = requests.Session()
        response = session.get(url, headers=headers)
        print(response)
        soup = BeautifulSoup(response.text, 'html.parser')
        jobs_on_page = extract_jobs_from_page(soup)
        all_jobs.extend(jobs_on_page)
        
        # Add a delay between requests to avoid being blocked
        time.sleep(random.uniform(1, 3))
        time.sleep(2)
    
    return pd.DataFrame(all_jobs)

# Fetch data for the first 3 pages of job listings (change 'q=developer' to fetch jobs for other positions)
url = "https://www.indeed.com/jobs?q=developer&l="  # Modify 'q' for different jobs or use an empty string to fetch all jobs
job_data = fetch_job_listings(url, pages=3)

# Save the data to a CSV file
job = job_data.to_csv('indeed_job_listings.csv')
print(job)
print("Data saved to indeed_job_listings.csv")
```
